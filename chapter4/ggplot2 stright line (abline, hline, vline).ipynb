```r
library(ggplot2)
```

```r
p <- ggplot(data = mtcars,
           mapping = aes(wt,mpg)) + 
        geom_point()
```

```r
p + geom_hline(yintercept = 20) + geom_vline(xintercept = 5)
```

```r
coef(lm(mpg ~ wt, data = mtcars))
```

```r
p + geom_abline(intercept = 37.285126167342, slope = -5)
```

```r
p + geom_smooth(method = "lm", se = FALSE)
```

```r
p <- ggplot(data = mtcars,
            mapping = aes(x = mpg, y =wt)) +
        geom_point() +
        facet_wrap(~ cyl)
```

```r
mean_wt <- data.frame(cyl = c(4,6,8), wt = c(2.28, 3.11, 4.00))
p + geom_hline(mapping = aes(yintercept = wt), data = mean_wt)
```

```r
ggplot(data = mtcars,
       mapping = aes(mpg, wt, colour = wt)) + 
    geom_point() + 
    geom_hline(mapping = aes(yintercept = wt, colour = wt), mean_wt) + 
    facet_wrap(~ cyl)
```
