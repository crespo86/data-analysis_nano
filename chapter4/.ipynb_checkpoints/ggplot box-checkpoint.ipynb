```r
library(ggplot2)
```

```r
p <- ggplot(data = mpg,
            mapping = aes(x = class, y = hwy))
p + geom_boxplot()
```

```r
p + geom_boxplot() + geom_jitter(width = 0.2)
```

```r
p + geom_boxplot() + coord_flip()
```

```r
p + geom_boxplot(notch = TRUE)
```

```r
p + geom_boxplot(varwidth = TRUE)
```

```r
p + geom_boxplot(fill = "white", colour = "#3366FF")
```

```r
p + geom_boxplot(outlier.colour = 'red', outlier.shape = 3)
```

```r
ggplot(data = diamonds,
       mapping = aes(x = carat, y = price)) + 
    geom_boxplot()
```

```r
ggplot(data = diamonds,
       mapping = aes(x = carat, y = price)) +
    geom_boxplot(mapping = aes(group = cut_width(carat, 0.25)))
```

```r
y <- rnorm(100)
```

```r
df <- data.frame(x = 1,
                 y0 = min(y),
                 y25 = quantile(y, 0.25),
                 y50 = median(y),
                 y75 = quantile(y,0.75),
                 y100 = max(y)
                )
```

```r
ggplot(data = df,
       mapping = aes(x = x)) +
    geom_boxplot(mapping = aes(ymin = y0, lower = y25, middle = y50, upper = y75, ymax = y100),
                 stat = 'identity')
```
