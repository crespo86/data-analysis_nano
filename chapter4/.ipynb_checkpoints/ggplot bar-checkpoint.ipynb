```r
g <- ggplot(data = mpg,
            mapping = aes(x = class))
```

```r
g + geom_bar()
```

```r
g + geom_bar(mapping = aes(weight = displ))
```

```r
df <- data.frame(trt = c("a","b","c"), outcome = c(2.3,1.9,3.2))
ggplot(data = df,
       mapping = aes(x= trt, y = outcome)) +
    geom_bar(stat = "identity")
```

```r
ggplot(data = df,
       mapping = aes(x = trt, y = outcome)) +
    geom_point()
```

```r
df <- data.frame(x = rep(c(2.9, 3.1, 4.5), c(5,10,4)))
```

```r
ggplot(data = df,
      mapping = aes(x = x)) +
    geom_bar()
```

```r
ggplot(data = df,
       mapping = aes(x = x)) + 
    geom_histogram(binwidth = 0.5)
```

```r
g + geom_bar(mapping = aes(fill = drv))
```

```r
g + geom_bar(mapping = aes(fill = drv),
            position = 'dodge')
```

```r
g + geom_bar(mapping = aes(fill = drv),
             position = 'fill')
```

```r
reorder_size <- function(x) {
                    factor(x, levels = names(sort(table(x))))
}
```

```r
ggplot(data = mpg,
       mapping = aes(x = reorder_size(class))) +
    geom_bar()
```
